### Determine Optimal Number of Clusters (Elbow Method)

### Apply K-Means Clustering

```python
# Assuming the elbow method suggests an optimal number of clusters, let's pick k=3 as a common practice.
# You can adjust this 'n_clusters' value based on the elbow plot from the previous step.
k = 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(scaled_df)

# Add the cluster labels to the original dataframe
df['Cluster'] = clusters

print(f"Number of data points per cluster (k={k}):")
display(df['Cluster'].value_counts().sort_index())
```
